```python
# Function to extract Job Title
def get_title(soup):
    try:
        title = soup.find("h1", attrs={"class":'text-primary pt-2 my-3 mb-4 mx-2 fs-2'})
        title_value = title.text
        title_string = title_value.strip()
    except AttributeError:
        title_string = ""
    return title_string

# Function to extract Job Description
def get_description(soup):
    try:
        description = soup.find("div", attrs={"class":'row p-3 pt-1 pb-3'}).text.strip()
    except AttributeError:
        description = ""
    return description

# Function to extract Job Requirement
def get_requirement(soup):
    requirement = ""
    try:
        requirement_tag = soup.find('dt', class_='col-sm-4', string='Experience Required:')
        
        if requirement_tag:
            require_tag = requirement_tag.find_next_sibling('dd', class_='col-sm-8')
        
            if require_tag:
                requirement = require_tag.text.strip()
            else:
                requirement = "not required"
    except AttributeError:
        requirement = ""
    return requirement

# Function to extract Job_type
def get_job_type(soup):
    job_type = ""
    try:
        job_type_tag = soup.find('dt', class_='col-sm-4', string='Contract Type:')
        
        if job_type_tag:
            contract_tag = job_type_tag.find_next_sibling('dd', class_='col-sm-8')
            
            if contract_tag:
                job_type = contract_tag.text.strip()  # Only get the text of the dd element
            else:
                job_type = "no contract"
    except AttributeError:
        job_type = ""
    return job_type
```

```python
# Scraping through different pages
# Function to extract Product Title
# ... (your function definitions go here) ...

if __name__ == '__main__':
    
    #add your headers
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/119.0', 'Accept-Language':'en-US, en;q=0.5'})
    
    #define your url
    URL = "https://ng.jobslin.com/job-offers/data?t=254&page="
    
    #create an empty dataframe to store details
    d = {"title":[], "description":[], "requirement":[], "job_type":[]}

    # Loop over the pages
    for page in range(1, 23):  # Corrected indentation here
        webpage = requests.get(URL + str(page), headers=HEADERS)
        soup = BeautifulSoup(webpage.content, "html.parser")
        links = soup.find_all("a", attrs={'class':'ink-secondary'})

        links_list = []  # Define links_list here

        # Loop for extracting links from link list
        for link in links:
            links_list.append(link.get('href'))

        # Loop for extracting all job details from each link
        for link in links_list:
            new_webpage = requests.get('https://ng.jobslin.com' + link, headers=HEADERS)
            new_soup = BeautifulSoup(new_webpage.content, "html.parser")

            # Functions to display all job details
            d['title'].append(get_title(new_soup))
            d['description'].append(get_description(new_soup))
            d['requirement'].append(get_requirement(new_soup))
            d['job_type'].append(get_job_type(new_soup))

    df = pd.DataFrame.from_dict(d)
    df['title'].replace('', np.nan, inplace=True)
    df = df.dropna(subset=['title'])
    df.to_csv("JobsLin_data.csv", header=True, index=False)
```
